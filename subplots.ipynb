```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.figure()

plt.subplot(1, 2, 1) # rows, columns, current axis is the first subplot axes, indices start at ONE!
# these numbers also can be only one parameter, 121 in this case

linear_data = np.arange(1, 9)
plt.plot(linear_data, "-o");
```

```python
plt.subplot(1, 2, 1)
plt.plot(linear_data, "-o")

plt.subplot(1, 2, 2) # the last number is 2 so that these are "different" plots
exponential_data = linear_data ** 2
plt.plot(exponential_data, "-o");
```

```python
# when we know we are going to share some axis, we need to create a variable for the subplot
axis1 = plt.subplot(1, 2, 1)
plt.plot(linear_data, "-o")

axis2 = plt.subplot(1, 2, 2, sharey=axis1) # the "second" plot will share the y axis with the "first" one
plt.plot(exponential_data, "-o");
```

```python
# creating a 3x3 grid of subplots using tuple unpacking
fig, ((ax1, ax2, ax3), (ax4, ax5, ax6), (ax7, ax8, ax9)) = plt.subplots(3, 3, sharex=True, sharey=True) # subplotS
ax5.plot(linear_data, "-"); # only the 5th subplot (the middle one) will display data
```

```python
fig = plt.gcf() # creating a new figure

for i in range(1, 7):
    if (i != 3 and i != 5): # do not plot the 3rd and 5th subplots
        axis = fig.add_subplot(2, 3, i) # position of the subplot depends on its number
        axis.text(0.5, 0.2, str((i)), fontsize=18, ha="center") # writing the number of the subplot in it
# since the axis were not shared, each subplot has its own x and y values and axis
```

## Testing with the Iris csv file

```python
df = pd.read_csv("iris.csv")
```

```python
columns = ["SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalWidthCm"] # the other columns tell about the species and the ID

fig, axs = plt.subplots(len(columns), len(columns), figsize=(10,10)) # creating a 4x4 subplots grid
# axs will be a matrix containing the axis of each subplot

for i in range(len(columns)):
    for j in range(len(columns)): # iterating through each subplot
        axs[i, j].scatter(df[columns[j]], df[columns[i]], s=5) # smaller size for better visualization
        
        # cleaning the subplots
        axs[i, j].get_xaxis().set_visible(False)
        axs[i, j].get_yaxis().set_visible(False)

        if (i == len(columns) - 1): # last row
            axs[i, j].get_xaxis().set_visible(True) # displaying the x-axis
            axs[i, j].set_xlabel(columns[j]) # displyaing the name of the column

        if (j == 0): # first column
            axs[i, j].get_yaxis().set_visible(True) # displaying the y-axis
            axs[i, j].set_ylabel(columns[i]) # displaying the name of the row
```
