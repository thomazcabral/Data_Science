```python
import matplotlib.pyplot as plt
import numpy as np
from random import randint
```

```python
linear_data = np.array([1, 2, 3, 4, 5, 6, 7, 8]) # 1 to 8

x_values = range(len(linear_data)) # 0 to 7
plt.bar(x_values, linear_data, width=0.5); # x values, y values and width
```

```python
exponential_data = linear_data**2

new_x_values = []
for value in x_values:
    new_x_values.append(value + 0.3)

plt.bar(x_values, linear_data, width=0.3)
plt.bar(new_x_values, exponential_data, width=0.3, color="red");
```

```python
linear_error = [randint(1, 4) for i in range(len(linear_data))] # creating some margin of error to the bar chart

plt.bar(x_values, linear_data, width=0.5, yerr=linear_error);
```

```python
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Cumulative chart")
# the same thing as from other charts

plt.bar(x_values, linear_data, width=0.5, color="blue")
plt.bar(x_values, exponential_data, width=0.5, bottom=linear_data, color="red")
# this shows the cumulative data, starting with the exponential data
```

```python
plt.barh(x_values, linear_data, height=0.5, color="blue")
plt.barh(x_values, exponential_data, height=0.5, left=linear_data, color="red");
# in order to create a horizontal bar chart, some things need to be done:
# we have to call plt.barh instead of plt.bar
# use height instead of width and left instead of bottom
```
