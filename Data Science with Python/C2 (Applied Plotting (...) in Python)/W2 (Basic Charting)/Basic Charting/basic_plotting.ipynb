```python
import matplotlib.pyplot as plt
```

```python
plt.plot(3, 2) # doesnt show anything beacuse it should have a 3rd element, symbolizing how the data have to be rendered
```

```python
plt.plot(3, 2, '.') # the dot means that we want to see (3, 2) as a point
```

```python
plt.figure() # just creating the background

plt.plot(3, 2, 'o')

axes = plt.gca() # reference to the current axes

axes.axis([0, 6, 0, 10])
# minimum value for x, maximum value for x
# minimum value for y, maximum value for y
```

```python
plt.figure()

plt.plot(1.5, 1.5, 'o', color="red")
plt.plot(2, 2, 'x', color="blue")
plt.plot(2.5, 2.5, '.', color="#00FA15") # green
# colors can be named after their names and by their HEX number

# creating titles along the plot
plt.title("Testing colors and titles")
plt.xlabel("X-axis title")
plt.ylabel("Y-axis title")
```
