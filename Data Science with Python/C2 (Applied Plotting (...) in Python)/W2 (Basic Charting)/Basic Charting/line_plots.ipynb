```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
linear_data = np.array([1, 2, 3, 4, 5, 6, 7, 8])
exponential_data = linear_data**2

plt.figure()
plt.plot(linear_data, "-o", exponential_data, "-o"); # "-o": between the dots (o), there is a line (-)
# if the semicolon wasnt put, this was also going to be printed:
#[<matplotlib.lines.Line2D at 0x17e17881a90>,
#<matplotlib.lines.Line2D at 0x17e178b9bd0>]
```

```python
plt.plot(linear_data, "-o", exponential_data, "-o")
plt.plot([22, 44, 55], "--r");
# --r creates a dashed red line passing through these 3 points: (0, 22), (1, 44) and (2, 55)
```

```python
# creating labels
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Line plots")

plt.legend(["Linear data", "Exponential data", "Dashed line"])
plt.plot(linear_data, "-o", exponential_data, "-o")
plt.plot([22, 44, 55], "--r");
```

```python
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Line plots")

plt.legend(["Linear data", "Exponential data", "Dashed line"])
plt.plot(linear_data, "-o", exponential_data, "-o")

plt.gca().fill_between(range(len(linear_data)), # the range of the fill
                        linear_data, exponential_data, # the data to be considered
                        facecolor="green",
                        alpha=0.2); # the transparency of the color
```

```python
# now I'll get into dates

dates = np.arange("2024-08-01", "2024-08-09", dtype="datetime64[D]")
plt.plot(dates, linear_data, "-o", dates, exponential_data, "-o");
# if we dont put dates, it'll show 0 to 7, so we have to put the dates as a parameter
```

```python
dates = np.arange("2024-08-01", "2024-08-09", dtype="datetime64[D]")
plt.plot(dates, linear_data, "-o", dates, exponential_data, "-o")

x_axis = plt.gca().xaxis # getting the x-axis cause theyre overlapping, so we can rotate them

for label in x_axis.get_ticklabels():
    label.set_rotation(45) # rotating it 45° counter-clockwise
```

```python
dates = np.arange("2024-08-01", "2024-08-09", dtype="datetime64[D]")
plt.plot(dates, linear_data, "-o", dates, exponential_data, "-o")

x_axis = plt.gca().xaxis # getting the x-axis cause theyre overlapping, so we can rotate them

for label in x_axis.get_ticklabels():
    label.set_rotation(45) # rotating it 45° counter-clockwise

axes = plt.gca() # when it is not specified if it is the x or y axis, it is both
axes.set_xlabel("Dates")
axes.set_ylabel("Numbers")
axes.set_title("Linear and exponential data");
# we need to but this set when the plot already exists and we want to change the labels
# or we could just put plt.xlabel, ylabel and title before calling plt.plot()
```
