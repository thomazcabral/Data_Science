```python
%matplotlib widget

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
```

```python
n = 100
sample = np.random.randn(n) # some random numbers for a sample
```

```python
# function that will do the plotting (FuncAnimation constructor)
def update(curr): # curr is the current frame
    if (curr == n): # checking if the animation is at the last frame
        a.event_source.stop()
    plt.cla()
    bins = np.arange(-4, 4, 0.5) # -4, -3.5, -3, ...
    plt.hist(sample[:curr], bins=bins) # it will show 1 element, then 2, then 3 ...
    plt.axis([-4, 4, 0, 30]) # defining the axis

    # Labels
    plt.gca().set_title("Sample for the normal distribution")
    plt.gca().set_xlabel("Value")
    plt.gca().set_ylabel("Frequency")
    plt.annotate(f"n = {curr}", [3, 27]) # these numbers are the xy location

a = animation.FuncAnimation(plt.figure(), update, interval=100)
# figure to be drawn on, FuncAnimation used, period of time between each update to the image
plt.show()
```
