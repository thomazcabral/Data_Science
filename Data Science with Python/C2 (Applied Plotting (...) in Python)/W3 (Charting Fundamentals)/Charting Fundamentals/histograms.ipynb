```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, sharex=True)
axes = [ax1, ax2, ax3, ax4]

for i in range(0, len(axes)):
    sample_size = 10 ** (i + 1) # 10, 100...
    sample = np.random.normal(loc=0.0, scale=1.0, size=sample_size) # loc is for the "center" of the distribution and scale is for the standard deviation
    axes[i].hist(sample)
    axes[i].set_title(f"n = {sample_size}")
# the bigger the sample size, the more it looks like a normal distribution
# by default, it creates 10 bins, that is, 10 bars in this case
```

```python
# now adding more bins to it

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, sharex=True)
axes = [ax1, ax2, ax3, ax4]

for i in range(0, len(axes)):
    sample_size = 10 ** (i + 1)
    sample = np.random.normal(loc=0.0, scale=1.0, size=sample_size)
    axes[i].hist(sample, bins=100) # they are added here
    axes[i].set_title(f"n = {sample_size}")
# now we can see a big difference between each plot
```

## Using gridspec

```python
import matplotlib.gridspec as gridspec
```

```python
yvalues = np.random.normal(loc=0.0, scale=1.0, size=10000)
xvalues = np.random.random(size=10000)
plt.scatter(xvalues, yvalues);
# it is not clear what the distributions really are for each axis
```

```python
plt.figure(figsize=(10,10))
gspec = gridspec.GridSpec(3, 3) #3x3 grid

# using slicing instead of using the 3 parameters (indices start at 0 here)
top_histogram = plt.subplot(gspec[0, 1:]) 
side_histogram = plt.subplot(gspec[1:, 0])
lower_right = plt.subplot(gspec[1:, 1:])

# this gives us a sense of the layout of what is being created
```

```python
plt.figure(figsize=(10,10))
gspec = gridspec.GridSpec(3, 3) #3x3 grid

# using slicing instead of using the 3 parameters (indices start at 0 here)
top_histogram = plt.subplot(gspec[0, 1:]) 
side_histogram = plt.subplot(gspec[1:, 0])
lower_right = plt.subplot(gspec[1:, 1:])

top_histogram.hist(xvalues, bins=100)
side_histogram.hist(yvalues, bins=100, orientation="horizontal") #rotate it
lower_right.scatter(xvalues, yvalues);

# getting rid of the whitespace
for axes in [top_histogram, lower_right]:
    axes.set_xlim(0, 1)
for axes in [side_histogram, lower_right]:
    axes.set_ylim(-5, 5)
```

## Using histograms with csv files

```python
df = pd.read_csv("Idades_vencedores_oscar.csv")

fig, (ax1, ax2) = plt.subplots(1, 2, sharex=True, sharey=True)

women = df["Idades_atrizes"]
men = df["Idades_atores"]

ax1.hist(women, bins=12)
ax1.set_title("Age woman won the Oscar")

ax2.hist(men, bins=12)
ax2.set_title("Age men won the Oscar");
```
