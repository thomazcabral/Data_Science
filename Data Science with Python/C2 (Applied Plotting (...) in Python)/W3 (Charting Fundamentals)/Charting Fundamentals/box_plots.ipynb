```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
normal_sample = np.random.normal(loc=0.0, scale=1.0, size=10000) # normal distribution
random_sample = np.random.random(size=10000)
gamma_sample = np.random.gamma(2, size=10000) # gamma distribution, 2 is for the shape of the distribution

df = pd.DataFrame({"normal": normal_sample, "random": random_sample, "gamma": gamma_sample})
df.describe()
```

```python
plt.boxplot(df["normal"]);
```

### Boxplot explained:
- Outliers: Those balls
- Max and Min without outliers: The lines above and under the rectangle
- First Quartile (25%): The bottom limit of the rectangle
- Median (Second Quartile): Line in the rectangle
- Third Quartile (75%): The upper limit of the rectangle
- Interquartile Range: The range of the limits of the rectangle

```python
plt.boxplot([df["normal"], df["random"], df["gamma"]], whis=[0, 100]);
# thw whis parameter removes the outliers for better visualization
# the boxplot is a bit "ugly" because the axes were not normalized
# the gamma distributions has some high max values, now I'll analyze it separately
```

```python
plt.hist(df["gamma"], bins=100);
# now everything makes sense
```

```python
plt.figure(figsize=(9,9))

plt.boxplot([df["normal"], df["random"], df["gamma"]], whis=[0, 100])

axis2 = plt.gca().inset_axes([0, 0.6, 0.6, 0.4]) # numbers needed to add the histogram to the top left of the boxplot
axis2.hist(df["gamma"], bins=100)
axis2.yaxis.tick_right(); # relocating the histogram's y-axis to the right
```
